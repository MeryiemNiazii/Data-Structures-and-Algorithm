```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        if not self.head:
            self.head = Node(data)
        else:
            new_node = Node(data)
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")             #only for display
            current = current.next                      #loop works unyil it print null
        print("None")

    def remove(self, key):
        if self.head is None:
            return

        if self.head.data == key:
            self.head = self.head.next
            return

        current = self.head
        while current.next:
            if current.next.data == key:
                current.next = current.next.next
                return
            current = current.next

    def search(self, value):
        if self.head is None:
            print("Node doesn't exist")
            return

        current = self.head
        position = 1
        while current:
            if current.data == value:
                print(f"Node {value} is at position {position}")
                return
            current = current.next
            position += 1

        print(f"Node {value} doesn't exist")

    def count(self):
        current = self.head
        count = 0
        while current:
            count += 1
            current = current.next
        return count

    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev


# Driver Code
def singlylistmenu():
    sll = SinglyLinkedList()
    while True:
        print("-------SinglyLinkedListMenu-------")
        print("Press 1 for Add Node")
        print("Press 2 for Remove Node")
        print("Press 3 for Search a node")
        print("Press 4 for count number of nodes")
        print("Press 5 for display the list")
        print("Press 6 for Reverse the list")
        print("Press 7 for Quit")
        a = int(input("Enter : "))
        if a == 1:
            b = input("Node: ")
            sll.append(b)           #enter a node 
            sll.display()
        elif a == 2:
            c = input("Remove node with data: ")
            sll.remove(c)
            sll.display()
        elif a == 3:
            d = input("Search data: ")
            sll.search(d)
        elif a == 4:
            print(f"Total Nodes: {sll.count()}")
        elif a == 5:
            sll.display()
        elif a == 6:
            sll.reverse()
            sll.display()
        elif a == 7:
            break
        else:
            print("Invalid Entry!")
```

# Stack

```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        e=len(self.items)
        if e==0:          #empty
            print("Stack is empty!")
        else:
            print("Stack is not empty.")

    def push(self, item):
        self.items.append(item)


    def pop(self, item=None):
        if self.is_empty():
            raise IndexError("Pop from an empty stack")
        
        if item is None or item == "":
            return self.items.pop()
        
        try:
            self.items.remove(item)
        except ValueError:
           print(f"Item {item} not found in the stack")
        
        return item
    
    def peek(self):
        if self.is_empty():
            raise IndexError("Peek from an empty stack")
        return self.items[-1]

    def size(self):
        return len(self.items)

    def __str__(self):
        return str(self.items)

    def display(self):
        print("Stack: " + " -> ".join(map(str, reversed(self.items))))
    
    
    
   
# Driver Code
def stackmenu():
    stack = Stack()
    x=0
    while True:
        print("-------StackMenu-------")
        print("Press 1 for add an element in stack")
        print("Press 2 for remove an element from stack")
        print("Press 3 for stack size")
        print("Press 4 to check if stack is empty")
        print("Press 5 to check the top element")
        print("Press 6 for display the list")
        print("Press 7 for Quit")
        a = int(input("Enter : "))
        if a == 1:
                b=int(input("To add an element in stack: "))
                stack.push(b)
                stack.display()
                
        elif a == 2:
                print("(Not input anything will result the top element to be removed)")
                c=input("To remove an element in stack : ")
                stack.pop(c)
        elif a == 3:
                l=stack.size()
                print(l)
        elif a == 4:
                stack.is_empty()
                
        elif a == 5:
                p=stack.peek()
                print(f"The top element of the stack is {p}")
        elif a == 6:
                stack.display()
        elif a == 7:
                break
        else:
            print("Invalid Entry!")
```

# Queue

```python
import queue

class CustomQueue:
    def __init__(self):
        self.items = queue.Queue()

    def is_empty(self):
        if self.items.qsize() == 0:
            print("Queue is empty!")
        else:
            print("Queue is not empty.")

    def enqueue(self, item):
        self.items.put(item)

    def dequeue(self, item=None):
        if self.is_empty():
            print("Dequeue from an empty queue")
            return None
        
        if item is None or item == "":
            return self._dequeue_fifo()
        
        return self._remove_item(item)

    def _dequeue_fifo(self):      #top of stack delete
        temp_queue = queue.Queue()
        dequeued_item = None

        while not self.items.empty():
            item = self.items.get()
            if dequeued_item is None:
                dequeued_item = item
            else:
                temp_queue.put(item)

        self.items = temp_queue
        if dequeued_item is None:
            print("Dequeue from an empty queue")
            return None

        return dequeued_item

    def _remove_item(self, item):
        temp_queue = queue.Queue()
        found = False

        while not self.items.empty():
            current_item = self.items.get()
            if current_item == item and not found:
                found = True
            else:
                temp_queue.put(current_item)

        self.items = temp_queue
        if not found:
            print(f"Item {item} not found in the queue")
            return None
        
        return item

    def peek(self):       #top element 
        if self.is_empty():
            print("Peek from an empty queue")
            return None
        temp_queue = queue.Queue()
        front_item = None

        while not self.items.empty():
            item = self.items.get()
            if front_item is None:
                front_item = item
            temp_queue.put(item)

        self.items = temp_queue
        return front_item

    def size(self):
        return self.items.qsize()

    def display(self):
        temp_queue = queue.Queue()
        elements = []

        while not self.items.empty():
            item = self.items.get()
            elements.append(item)
            temp_queue.put(item)

        self.items = temp_queue
        print("Queue: " + " -> ".join(map(str, elements)))


def queuemenu():
    q = CustomQueue()
    while True:
        print("-------QueueMenu-------")
        print("Press 1 for add an element in queue")
        print("Press 2 for remove an element from queue")
        print("Press 3 for queue size")
        print("Press 4 to check if queue is empty")
        print("Press 5 to check the top element")
        print("Press 6 for display the list")
        print("Press 7 for Quit")
        a = int(input("Enter : "))
        if a == 1:
                b=int(input("To add an element in queue: "))
                q.enqueue(b)
                q.display()
                
        elif a == 2:
                c=int(input("To remove an element in queue : "))
                q.dequeue(c)
        elif a == 3:
                l=q.size()
                print(l)
        elif a == 4:
                q.is_empty()
                
        elif a == 5:
                p=q.peek()
                print(f"The top element of the stack is {p}")
        elif a == 6:
                q.display()
        elif a == 7:
                break
        else:
            print("Invalid Entry!")
```

# Tree

```python
class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, key):
        if self.root is None:
            self.root = TreeNode(key)
        else:
            self._insert(self.root, key)

    def _insert(self, node, key):
        if key < node.key:
            if node.left is None:
                node.left = TreeNode(key)
            else:
                self._insert(node.left, key)
        else:
            if node.right is None:
                node.right = TreeNode(key)
            else:
                self._insert(node.right, key)

    def search(self, key):
        return self._search(self.root, key)

    def _search(self, node, key):
        if node is None or node.key == key:
            return node
        if key < node.key:
            return self._search(node.left, key)
        else:
            return self._search(node.right, key)

    def delete(self, key):
        self.root, deleted = self._delete(self.root, key)
        if deleted:
            print(f"Deleted {key} from the tree")
        else:
            print(f"Key {key} not found in the tree")

    def _delete(self, node, key):
        if node is None:
            return node, False
        
        if key < node.key:
            node.left, deleted = self._delete(node.left, key)
        elif key > node.key:
            node.right, deleted = self._delete(node.right, key)
        else:
            # Node to be deleted found
            if node.left is None:
                return node.right, True
            elif node.right is None:
                return node.left, True

            min_larger_node = self._get_min(node.right)
            node.key = min_larger_node.key
            node.right, _ = self._delete(node.right, min_larger_node.key)
            return node, True
        
        return node, False

    def _get_min(self, node):
        while node.left is not None:
            node = node.left
        return node

    def in_order_traversal(self):
        result = []
        self._in_order_traversal(self.root, result)
        return result

    def _in_order_traversal(self, node, result):
        if node is not None:
            self._in_order_traversal(node.left, result)
            result.append(node.key)
            self._in_order_traversal(node.right, result)

    def display(self):
        elements = self.in_order_traversal()
        print("BST In-Order Traversal: " + " -> ".join(map(str, elements)))

# Driver Code
def treemenu():
    bst = BinarySearchTree()
    while True:
        print("-------TreeMenu-------")
        print("Press 1 for add node in tree")
        print("Press 2 for remove data from tree")
        print("Press 3 for search")
        print("Press 4 for display the tree")
        print("Press 5 for Quit")
        a = int(input("Enter : "))
        if a == 1:
                b=int(input("To add Node in Tree: "))
                bst.insert(b)
                bst.display()
                
        elif a == 2:
                c=int(input("To remove data from tree : "))
                bst.delete(c)
        elif a == 3:
                d=input("Search : ")
                node = bst.search(d)
                if node:
                    print(f"Found {node.key} in the tree")
                else:
                    print(f"Key {d} not found in the tree")
        elif a == 4:
                bst.display()
        elif a == 5:
                break
        else:
            print("Invalid Entry!")
```

# Graph

```python
import heapq
from collections import deque

class Graph:
    def __init__(self):
        self.nodes = set()
        self.edges = {}
        self.distances = {}

    def add_node(self, value):
        self.nodes.add(value)
        self.edges[value] = []

    def add_edge(self, from_node, to_node, distance):
        self.edges[from_node].append(to_node)
        self.edges[to_node].append(from_node)  # assuming undirected graph
        self.distances[(from_node, to_node)] = distance
        self.distances[(to_node, from_node)] = distance

def dijkstra(graph, initial):
    visited = {initial: 0}
    path = {}

    nodes = set(graph.nodes)

    while nodes:
        min_node = None
        for node in nodes:
            if node in visited:
                if min_node is None:
                    min_node = node
                elif visited[node] < visited[min_node]:
                    min_node = node

        if min_node is None:
            break

        nodes.remove(min_node)
        current_weight = visited[min_node]

        for edge in graph.edges[min_node]:
            weight = current_weight + graph.distances[(min_node, edge)]
            if edge not in visited or weight < visited[edge]:
                visited[edge] = weight
                path[edge] = min_node

    return visited, path

def shortest_path(graph, origin, destination):
    visited, paths = dijkstra(graph, origin)
    full_path = deque()
    _destination = paths.get(destination)

    if _destination is None:
        return None, []

    while _destination != origin:
        full_path.appendleft(_destination)
        _destination = paths[_destination]

    full_path.appendleft(origin)
    full_path.append(destination)

    return visited[destination], list(full_path)

# Interactive menu
def graphmenu():
    graph = Graph()
    
    while True:
        print("-------Graph Menu-------")
        print("Press 1 to add a node to the graph")
        print("Press 2 to add an edge to the graph")
        print("Press 3 to find the shortest path")
        print("Press 4 to display the graph")
        print("Press 5 to Quit")
        choice = input("Enter your choice: ").strip()

        if choice == "1":
            node = input("Enter the node value to add: ").strip()
            graph.add_node(node)
            print(f"Node '{node}' added to the graph.")
        elif choice == "2":
            from_node = input("Enter the starting node of the edge: ").strip()
            to_node = input("Enter the ending node of the edge: ").strip()
            distance = int(input("Enter the distance between the nodes: ").strip())
            graph.add_edge(from_node, to_node, distance)
            print(f"Edge added from '{from_node}' to '{to_node}' with distance {distance}.")
        elif choice == "3":
            origin = input("Enter the starting node: ").strip()
            destination = input("Enter the destination node: ").strip()
            distance, path = shortest_path(graph, origin, destination)
            if path:
                print(f"The shortest path from '{origin}' to '{destination}' is {path} with total distance {distance}.")
            else:
                print(f"No path found from '{origin}' to '{destination}'.")
        elif choice == "4":
            print("Graph nodes:", graph.nodes)
            print("Graph edges:")
            for edge, distance in graph.distances.items():
                print(f"  {edge[0]} -> {edge[1]} with distance {distance}")
        elif choice == "5":
            break
        else:
            print("Invalid choice. Please try again.")
```

# Sorting Algorithm

```python
class SortingAlgorithms:
    def _init_(self):
        self.arr = []

    def set_list(self, arr):
        self.arr = arr

    def get_list(self):
        return self.arr

    def bubble_sort(self):
        arr = self.arr.copy()
        n = len(arr)
        for i in range(n):
            for j in range(0, n - i - 1):
                if arr[j] > arr[j + 1]:
                    arr[j], arr[j + 1] = arr[j + 1], arr[j]
        return arr

  
    def input_list(self):
        arr = []
        print("Enter the list of numbers one by one (enter 'done' to finish):")
        while True:
            num = input()
            if num.lower() == 'done':
                self.set_list(arr)
                print(f"List entered: {self.get_list()}")
                return
            try:
                arr.append(int(num))
            except ValueError:
                print("Invalid input. Please enter an integer.")


def sortmenu():
    sorting = SortingAlgorithms()
    while True:
        print("-------Sorting Menu-------")
        print("Press 1 to enter a list of numbers")
        print("Press 2 to sort the list using Bubble Sort")
        print("Press 3 to display the list")
        print("Press 4 to Quit")
        choice = input("Enter your choice: ")

        if choice == "1":
            sorting.input_list()
        elif choice == "2":
            sorted_arr = sorting.bubble_sort()
            print(f"List sorted using Bubble Sort: {sorted_arr}")
        elif choice == "3":
            print(f"Current list: {sorting.get_list()}")
        elif choice == "4":
            break
        else:
            print("Invalid choice. Please try again.")

```
